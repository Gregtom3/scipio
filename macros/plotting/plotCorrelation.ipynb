```python
import ROOT
import numpy as np
```

```python
def get_avg_ybin(h2d):
    #calculate the average ybin for each xbin
    x=[]
    y=[]
    avg_ybin = []
    for xbin in range(h2d.GetNbinsX()):
        ybin_sum = 0.
        count = 0.
        for ybin in range(h2d.GetNbinsY()):
            ybin_sum += h2d.GetBinContent(xbin+1, ybin+1)*ybin
            count += h2d.GetBinContent(xbin+1, ybin+1)
        if(count==0):
            continue
        avg_ybin.append(ybin_sum/count)
        x.append(h2d.GetXaxis().GetBinCenter(xbin+1))
        y.append(h2d.GetYaxis().GetBinCenter(int(np.round(avg_ybin[-1]))))
        
    return len(x),x,y
```

```python
# Obtain the TChain for Fall 2018 Inbending Data

import os
tmc=ROOT.TChain("dihadron")
prefix = "/volatile/clas12/users/gmat/clas12analysis.sidis.data/rga/ML/projects/piminuspi0_prelim/catboost/postprocess"
files = os.listdir(prefix)
for file in files:
    if("MC" in file):
        continue
    run=int(file.split(".")[0].split("_")[1])
    if(run<=5332):
        tmc.Add(prefix+f"/nSidis_{run}.root")
```

```python
def create_corr_plot(tmc,branchName,ytitle,xmin,xmax):
    h2=ROOT.TH2F("h2",";M_{#gamma#gamma} [GeV];"+ytitle,100,0.06,0.25,100,xmin,xmax)
    ROOT.gStyle.SetNumberContours(300)
    ROOT.gStyle.SetPalette(ROOT.kRainBow)
    c=ROOT.TCanvas("c","c",500,500)
    c.SetRightMargin(0.12)
    c.SetBottomMargin(0.12)
    c.SetLeftMargin(0.12)
    c.SetTopMargin(0.12)
    tmc.Draw(f"{branchName}:Mgg>>h2","prob_g1>0.9&&prob_g2>0.9","goff")
    h2.GetXaxis().SetNdivisions(508)
    h2.GetXaxis().SetTitleSize(0.05)
    h2.GetYaxis().SetTitleSize(0.05)
    ROOT.gStyle.SetOptStat(0)
    c.SetLogz(1)

    N,x,y = get_avg_ybin(h2)
    graph=ROOT.TGraph(N,np.array(x),np.array(y))
    graph.SetLineWidth(3)
    graph.SetLineColor(6)
    #graph.Draw("l same")

    pavetext = ROOT.TPaveText(0.45, 0.73, 0.85, 0.85, "brNDC")
    pavetext.AddText("Fall 2018 Inbending Sample")
    pavetext.AddText("#bf{10.6#it{GeV}} e+p#rightarrow e'+#gamma#gamma+#pi^{-}+X")
    pavetext.AddText("CatBoost p(#gamma)>0.9")
    pavetext.SetTextSize(0.03)
    pavetext.SetTextFont(42)
    pavetext.SetLineColor(1)
    pavetext.SetLineWidth(1)
    #pavetext.Draw("same")

    pavetext2 = ROOT.TPaveText(0.16,0.8,0.33,0.85,"brNDC")
    pavetext2.AddText("#bf{corr}="+"{}".format(np.round(h2.GetCorrelationFactor(),3)))
    pavetext2.SetTextSize(0.03)
    pavetext2.SetTextFont(42)
    #pavetext2.Draw("same")

    return c,h2,graph,pavetext,pavetext2
```

```python
c,h2,graph,pavetext,pavetext2=create_corr_plot(tmc,"x","x",0,1)

h2.Draw("colz")
graph.Draw("l same")
pavetext.Draw("same")
pavetext2.Draw("same")
c.Draw()
```
