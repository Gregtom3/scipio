```python
PROJECT_NAME="pipluspi0_noresonance_tiny"
MLmethod="catboost"
```

```python
bin_dirs, outdir = get_dirs(PROJECT_NAME,MLmethod)
histnames = [key for key in pltfmt.keys()]
DATATYPE=["all", "Fall2018_inbending", "Fall2018_outbending", "Spring2019_inbending"]
MCTYPE=["all", "MC_inbending", "MC_outbending", "MC_inbending"]

for d in bin_dirs:
    full_plots_data, abbrev_plots_data = collect_many(d,
                                                      "nSidis",
                                                      DATATYPE)
    full_plots_MC, abbrev_plots_MC = collect_many(d,
                                                  "MC",
                                                  MCTYPE)
    
    for datatype,mctype in zip(DATATYPE,MCTYPE):
        # mkdir
        if(not os.path.exists(f"{outdir}/{datatype}/")):
                os.mkdir(f"{outdir}/{datatype}")
        for histname in histnames:
                for B in [False,True]: # False first, or else the dodiff histogram scaling will mess things up
                    for R in [False, True]: # Determine whether to rescale the y axis for all plots
                            make_binned_plots(params_data=full_plots_data,
                                              params_MC=full_plots_MC,
                                              histname=histname,
                                              histname_data=histname+"_"+datatype,
                                              histname_MC=histname+"_"+mctype,
                                              OUTDIR=outdir+"/"+datatype,
                                              skipEdges=False, # Since params is full
                                              dodiff=B,
                                              rescale_y=R)
                            make_binned_plots(params_data=abbrev_plots_data,
                                              params_MC=abbrev_plots_MC,
                                              histname=histname,
                                              histname_data=histname+"_"+datatype,
                                              histname_MC=histname+"_"+mctype,
                                              OUTDIR=outdir+"/"+datatype,
                                              skipEdges=True, # Since params is abbreviated
                                              dodiff=B,
                                              rescale_y=R)
```
