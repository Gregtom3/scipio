```python

bin_dirs, outdir = get_dirs(PROJECT_NAME,MLmethod)

histnames = [key for key in pltfmt.keys()]

DATATYPE=["all", "Fall2018_inbending", "Fall2018_outbending", "Spring2019_inbending"]
CUTDATA=["", "run>=5032 && run<=5332", "run>=5333 && run<=5666" , "run>=6616 && run<=6783"]
CUTMC=["","run==-11","run==11","run==-11"]
for b in [True,False]:
    for d in bin_dirs:
        for datatype,cutdata,cutmc in zip(DATATYPE,CUTDATA,CUTMC):
            plots_data = collect(input_dir = d,
                         skipEdges=b,
                         datatype="nSidis",
                         cut=cutdata)
            plots_MC = collect(input_dir = d,
                         skipEdges=b,
                         datatype="MC",
                         cut=cutmc)
            # mkdir
            if(not os.path.exists(f"{outdir}/{datatype}/")):
                    os.mkdir(f"{outdir}/{datatype}")
            for histname in histnames:
                for B in [False,True]: # False first, or else the dodiff histogram scaling will mess things up
                    make_binned_plots(params_data=plots_data,
                                      params_MC=plots_MC,
                                      histname=histname,
                                      outdir=outdir+"/"+datatype,
                                      skipEdges=b,
                                      dodiff=B)
```
